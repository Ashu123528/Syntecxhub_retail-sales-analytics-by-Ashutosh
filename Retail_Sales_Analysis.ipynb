### Company-Level Exploratory Data Analysis Project

**Objective:** Perform univariate, bivariate, and multivariate EDA on retail sales data to generate actionable business insights.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10,5)
```

## 2. Load Dataset

```python
df = pd.read_csv('retail_sales_data.csv')
df.head()
```

## 3. Data Understanding

```python
df.info()
```

```python
df.describe()
```

## 4. Data Cleaning

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df.isnull().sum()
```

## 5. Univariate Analysis
Understanding individual variable distributions.

```python
sns.histplot(df['sales'], bins=30, kde=True)
plt.title('Sales Distribution')
plt.show()
```

## 6. Bivariate Analysis
Relationship between two variables.

```python
sns.boxplot(x='category', y='profit', data=df)
plt.title('Profit Distribution by Category')
plt.show()
```

## 7. Multivariate Analysis
Combined impact of region and category.

```python
pivot = df.pivot_table(values='sales', index='region', columns='category', aggfunc='mean')
pivot.plot(kind='bar')
plt.title('Average Sales by Region and Category')
plt.ylabel('Average Sales')
plt.show()
```

## 8. Key Business Insights
1. West and North regions generate higher average sales.
2. Technology category shows higher profit variability.
3. Some regions have high sales but low profits, indicating cost optimization opportunities.

## 9. Conclusion
This analysis demonstrates how EDA helps uncover business patterns and supports data-driven decision-making.
